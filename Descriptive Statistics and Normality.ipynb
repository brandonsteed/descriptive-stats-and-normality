```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
poissonvar = np.random.poisson(2, 500)
plt.hist(poissonvar)
plt.axvline(poissonvar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(poissonvar.mean() - poissonvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(poissonvar.mean() + poissonvar.std(), color='r', linestyle='dashed', linewidth=2)
```

Median is probably more useful than mean and standard deviation in describing this variable.

```python
chisqvar = np.random.chisquare(2, 100)
plt.hist(chisqvar)
plt.axvline(chisqvar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(chisqvar.mean() - chisqvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(chisqvar.mean() + chisqvar.std(), color='r', linestyle='dashed', linewidth=2)
```

Mean and standard deviation are not very useful in describing this variable.

```python
geovar = np.random.geometric(p=.25, size=100)
plt.hist(geovar)
plt.axvline(geovar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(geovar.mean() - geovar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(geovar.mean() + geovar.std(), color='r', linestyle='dashed', linewidth=2)
```

Mean and standard deviation are not very useful in describing this variable.

```python
expovar = np.random.exponential(2, 100)
plt.hist(expovar)
plt.axvline(expovar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(expovar.mean() - expovar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(expovar.mean() + expovar.std(), color='r', linestyle='dashed', linewidth=2)
```

Mean and standard deviation are not very useful in describing this variable.

```python
hypergeometricvar = np.random.hypergeometric(150, 100, 50, 500)
plt.hist(hypergeometricvar)
plt.axvline(hypergeometricvar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(hypergeometricvar.mean() - hypergeometricvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(hypergeometricvar.mean() + hypergeometricvar.std(), color='r', linestyle='dashed', linewidth=2)
```

Mean and standard deviation seem to be useful in describing this variable.

```python
binomialvar = np.random.binomial(50, .4, 500)
plt.hist(binomialvar)
plt.axvline(binomialvar.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(binomialvar.mean() - binomialvar.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(binomialvar.mean() + binomialvar.std(), color='r', linestyle='dashed', linewidth=2)
```

Mean and standard deviation seem to be useful in describing this variable.

The closer to a normal distribution, the more useful mean and standard deviation are in describing the data.  Mean and standard deviation are not very useful in describing distributions that are highly skewed.

```python
norm1 = np.random.normal(5, .5, 1000)
plt.hist(norm1)
norm2 = np.random.normal(10, 1, 1000)
plt.hist(norm2)
```

```python
norm3 = norm1 + norm2
plt.hist(norm3)
plt.axvline(norm3.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(norm3.mean() - norm3.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(norm3.mean() + norm3.std(), color='r', linestyle='dashed', linewidth=2)
print(norm3.mean())
print(norm3.std())
```
